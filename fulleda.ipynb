```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("realistic.csv")
```

```python
df = df.drop_duplicates()
df = df.convert_dtypes()
```

```python
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
```

```python
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)
```

```python
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
print("\n===== DATA OVERVIEW =====")
print(df.head())

print("\n===== SUMMARY =====")
print(df.describe())
```

```python
plt.figure()
plt.hist(df['Age'])
plt.title("Age Distribution")
plt.show()
```

```python
print("Insight 1: Most customers fall in middle age group.")
```

```python
plt.figure()
plt.hist(df['Salary'])
plt.title("Salary Distribution")
plt.show()

print("Insight 2: Salary distribution shows variation with some high earners.")
```

```python
plt.figure()
plt.hist(df['Purchases'])
plt.title("Purchases Distribution")
plt.show()

print("Insight 3: Purchases vary widely among customers.")
```

```python
plt.figure()
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.show()

print("Insight 4: Gender distribution shows dominant customer segment.")
```

```python
plt.figure()
df['City'].value_counts().plot(kind='bar')
plt.title("City Distribution")
plt.show()

print("Insight 5: Certain cities contribute more customers.")
```

```python
plt.figure()
plt.scatter(df['Age'], df['Salary'])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()

print("Insight 6: No strong linear relationship between age and salary.")
```

```python
plt.figure()
plt.scatter(df['Age'], df['Purchases'])
plt.title("Age vs Purchases")
plt.show()

print("Insight 7: Purchases are not strongly dependent on age.")
```

```python
plt.figure()
plt.scatter(df['Salary'], df['Purchases'])
plt.title("Salary vs Purchases")
plt.show()

print("Insight 8: Higher salary does not always mean more purchases.")
```

```python
plt.figure()
plt.boxplot(df['Salary'])
plt.title("Salary Boxplot")
plt.show()

print("Insight 9: Salary contains extreme values indicating outliers.")
```

```python
corr = df[num_cols].corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(num_cols)), num_cols, rotation=45)
plt.yticks(range(len(num_cols)), num_cols)
plt.title("Correlation Heatmap")
plt.show()

print("Insight 10: Features have weak correlation, indicating independence.")
```
